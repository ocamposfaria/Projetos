## Get coordinates

```python
def get_coordinates(row):
    var = client.find_place(input=row['Google Maps'], input_type='textquery', fields=['geometry'], language='portuguese')
    coordinates = (var['candidates'][0]['geometry']['location']['lat'], var['candidates'][0]['geometry']['location']['lng'])
    return coordinates
```

```python
df['coordinates'] = df.apply(get_coordinates, axis=1)
```
