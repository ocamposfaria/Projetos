## 2. Formulação do Modelo

```python
from IPython.display import Image
Image(r"C:\Users\ocamp\Desktop\Captura de tela 2022-07-17 181009.jpg")
```

```python
# CD: nós dos centros de distribuição (CDs);
# CDR: nós de unidades;

# Xij: variável binária para alocar a demanda do j-ésimo CDR ao i-ésimo CD;
# Yi: variável binária para abrir o i-ésimo CD;

# p é input de quantidade de CDs abertos (conforme p-mediana)
```

```python
prob = LpProblem("THE_P_MEDIAN_PROBLEM", LpMinimize)
```

```python
p = 6
```

```python
x = LpVariable.dicts('X_%s_%s', (CDs,unidades), 
    cat = 'Binary', 
    lowBound = 0, 
    upBound = 1)
```

```python
y = LpVariable.dicts('Y_%s', (CDs),
    cat = 'Binary', 
    lowBound = 0, 
    upBound = 1)
```

```python
prob += sum(x[i][j]*df[i][j] for i in CDs for j in unidades), "Distância total entre CDs e unidades"
```

```python
#prob += sum(x[i][j] for i in CDs for j in unidades) == 68
```

```python
for j in unidades:
        prob += sum(x[i][j] for i in CDs) == 1
```

```python
prob += lpSum(y[i] for i in CDs) == p
```

```python
for i in CDs:
    for  j in unidades:
        prob += x[i][j]<= y[i]
```

```python
prob
```

```python
prob.solve()
```

### Resultado da execução do modelo:

```python
for v in prob.variables():
    print(v.name , '=', v.varValue)
    #if v.varValue == 1:
    #    print(v.name , '=', v.varValue)
```
