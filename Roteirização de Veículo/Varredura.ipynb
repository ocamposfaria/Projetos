## 3. Calculando os ângulos

```python
Image(url= 'https://static.preparaenem.com/conteudo/images/calculo-do-angulo-entre-dois-vetores.jpg')
```

```python
x2 = -15.860712-lat_deposito
y2 = -47.872009-lng_deposito

nova_coluna = []

for i in df_coord.index:
    x1 = df_coord.loc[i][0]-lat_deposito
    y1 = df_coord.loc[i][1]-lng_deposito
    angulo = math.degrees(math.acos((x1*x2+y1*y2)/((math.sqrt(x1**2+y1**2))*(math.sqrt(x2**2+y2**2)))))
    nova_coluna.append(angulo)
    
df_coord['ângulo θ depósito'] = nova_coluna
df_coord = df_coord.sort_values(by=['ângulo θ depósito'], ascending = True)
df_coord
```

## 4. Aplicando a varredura

```python
contador = 0
roteiros = []
roteiro_embriao = [df_coord.index[0]]
roteiros.append(roteiro_embriao)

df_coord = df_coord.drop(roteiro_embriao[0])

for i in df_coord.index:
    contador = 0 + 1
    if i != roteiro_embriao:
        novo_roteiro = roteiros[len(roteiros)-1]
        novo_roteiro = novo_roteiro + [i]

        #CHECK RESTRICTIONS

        distancia_percorrida = 0
        tempo_descarga = 0
        peso_total = 0

        novo_roteiro_com_deposito = [deposito] + novo_roteiro + [deposito]

        for x in range(len(novo_roteiro)-1):
            distancia_percorrida = distancia_percorrida + map_client.distance_matrix(novo_roteiro_com_deposito[x],novo_roteiro_com_deposito[x+1], mode='driving')["rows"][0]['elements'][0]['distance']['value']

        for y in novo_roteiro:
            if y != deposito:
                peso_total = peso_total + restriçoes.loc[y][0]
                tempo_descarga = tempo_descarga + restriçoes.loc[y][1]

        tempo_total = distancia_percorrida/1000 + tempo_descarga

        if tempo_total < 6*60 and peso_total < 1600:
            del roteiros[len(roteiros)-1]
            roteiros = roteiros + [novo_roteiro]
        else:
            roteiros = roteiros + [[i]]
        
roteiros
```

#### 4.1. Indicadores

```python
contador = 0

for i in roteiros:
    
    distancia_percorrida = 0
    tempo_total = 0 
    tempo_descarga = 0
    peso_total = 0
    contador = contador + 1
    
    i_com_deposito = [deposito] + i + [deposito]
    for x in range(len(i)-1):
        distancia_percorrida = distancia_percorrida + map_client.distance_matrix(i_com_deposito[x],i_com_deposito[x+1], mode='driving')["rows"][0]['elements'][0]['distance']['value']

    for y in i:
        if y != deposito:
            peso_total = peso_total + restriçoes.loc[y][0]
            tempo_descarga = tempo_descarga + restriçoes.loc[y][1]

    tempo_total = distancia_percorrida/1000 + tempo_descarga
    

    print('Para o roteiro ' + str(contador) + ':')
    #print('A distância total percorrida foi de ' + str(distancia_percorrida/1000) + ' quilômetros')
    print('O peso total é de: ' + str(peso_total) + ' quilogramas')
    print('O tempo total (transporte e descarga) é de: ' + str((tempo_total)/60) + ' horas')
    print('')
```
